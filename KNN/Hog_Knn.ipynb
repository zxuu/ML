```python
f=open('./flowers/test.txt', encoding="utf-8")
test_data = f.readlines()  # 直接将文件中按行读到list里，效果与方法2一样
f.close()
```

```python
i = 0
test_label = []
for _ in test_data:
    elements = re.split(r'[-\s,;!?]+', _)
    test_data[i] = elements[0]
    test_label.extend(elements[1])
    i += 1
i = 0
```

```python
test_data_hog = []
j = 0
for test_data_path in test_data:
    img = cv2.imread('./flowers/'+test_data_path, cv2.IMREAD_GRAYSCALE)
    img2 = cv2.resize(img, (224, 224))
#    if not img is None:
#        img = cv2.imread('./Result/'+test_data_path, cv2.IMREAD_GRAYSCALE)
    feature = ft.hog(img2,orientations=6,pixels_per_cell=[20,20],cells_per_block=[2,2],visualize=False)
    test_data_hog.append(feature)
    j += 1
```

```python
test_data = [_.tolist() for _ in test_data_hog]
```

```python
import matplotlib.pyplot as plt
ks = np.arange(2,30,2)
scores = []
for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(data,train_label)
    score = knn.score(test_data,test_label)
    scores.append(score)
scores = np.array(scores)
plt.plot(ks,scores)
plt.xlabel('k')
plt.ylabel('score')
```
